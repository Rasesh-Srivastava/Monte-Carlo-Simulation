```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 08

# Question 1
import numpy as np
def StraInt(PrStr, n2):
    z = 2.576
    std_error_stratified = np.sqrt(
        (PrStr * (1 - PrStr)) / n2)
    confidence_interval_stratified = (
        PrStr - z * std_error_stratified, PrStr + z * std_error_stratified)
    print(f"99% Confidence Interval (Stratification): {confidence_interval_stratified}")
def MC_int(PrMC, n1):
    z = 2.576
    std_error_mc = np.sqrt((PrMC * (1 - PrMC)) / n1)
    confidence_interval_mc = (PrMC - z * std_error_mc,
                              PrMC + z * std_error_mc)
    print(f"99% Confidence Interval (Monte Carlo): {confidence_interval_mc}")

n1 = 100
n2 = 10000
Lamb = 2.9
shape_k = 0.8
scale_sigma = 3

def MyWeibull(k, scale, size):
    return np.random.weibull(k, size) * scale

def MCPro(n):
    rainfall = np.zeros(n)
    for i in range(n):
        num_storms = np.random.poisson(Lamb)
        storm_depths = MyWeibull(shape_k, scale_sigma, num_storms)
        total_rainfall = np.sum(storm_depths)
        rainfall[i] = total_rainfall
    return np.mean(rainfall < 5)


def stratification_probability(n):
    strata = [0, 1, 2, 3, 4, 5]
    num_strata = len(strata)
    probs = np.zeros(num_strata)
    for i in range(num_strata):
        S = strata[i]
        prob_S = 0
        for _ in range(n):
            num_storms = np.random.poisson(Lamb)
            storm_depths = MyWeibull(shape_k, scale_sigma, num_storms)
            total_rainfall = np.sum(storm_depths)
            if S == num_storms:
                prob_S += (total_rainfall < 5)
        probs[i] = prob_S / n
    return np.sum(probs)


MC1 = MCPro(n1)
print(f"Monte Carlo Probability for n={n1}: {MC1:.4f}")
MC_int(MC1,100)

Stra1 = stratification_probability(n1)
print(
    f"Stratification Method Probability for n={n1}: {Stra1:.4f}")
StraInt(Stra1,100)

MC2 = MCPro(n2)
print(f"Monte Carlo Probability for n={n2}: {MC2:.4f}")
MC_int(MC2,10000)

Stra2 = stratification_probability(n2)
print(f"Stratification Probability for n={n2}: {Stra2:.4f}")
StraInt(Stra2,10000)
```

```python
# Question 2
import numpy as np 
import matplotlib.pyplot as graphPlotter  
from collections import OrderedDict
print("The Linear Congruence Generator is: ")
print("x_i+1 = (a * x_i + b) mod m")
print("u_i+1 = x_i+1 / m")
print()
print("Question 2 :")
print()
def SequenceGeneratorHelper(a, b, m, seed):
    x = seed
    ui = list()
    xi = list()

    sequence = dict()
    j=0
    while(j<=19):
        sequence[j] = 0
        j+=1

    while True:
        x = (a * x + b) % m
        u = x/m
        if x in xi:
            break
        ui.append(u)
        xi.append(x)

        key = round(u*20) % 20
        sequence[key] = sequence[key] + 1
    return sequence

givenRanges = ['0 - 0.05', '0.05 - 0.10', '0.10 - 0.15', '0.15 - 0.20', '0.20 - 0.25', '0.25 - 0.30', '0.30 - 0.35', '0.35 - 0.40', '0.40 - 0.45', '0.45 - 0.50', '0.50 - 0.55', '0.55 - 0.60', '0.60 - 0.65', '0.65 - 0.70', '0.70 - 0.75', '0.75 - 0.80', '0.80 - 0.85', ' 0.85 - 0.90', '0.90 - 0.95', '0.95 - 1']
seedValues = [1,2,3,4,6]

for Seed in seedValues:
    sequence1 = SequenceGeneratorHelper(1597, 1, 244944, Seed)
    sequence1 = OrderedDict(sorted(sequence1.items())) 
    temporar = list(sequence1.keys())
    sequence1 = dict((round(key*0.05, 2), value) for (key, value) in sequence1.items())

    x = list(sequence1.keys())
    y = list(sequence1.values())

    graphPlotter.figure(figsize=(18, 5))
    width = np.min(np.diff(temporar))/3
    graphPlotter.bar(temporar - 0.5*width, y, color='red',  ec='black',width=0.3, label='a = 1597') 

    sequence2 = SequenceGeneratorHelper(51749, 1, 244944, Seed)
    sequence2 = OrderedDict(sorted(sequence2.items())) 
    temporar = list(sequence2.keys())
    sequence2 = dict((round(key*0.05, 2), value) for (key, value) in sequence2.items())

    x = list(sequence2.keys())
    y = list(sequence2.values())

    graphPlotter.bar(temporar + 0.5*width, y, color='cyan',  ec='black', width=0.3, label='a = 51749') 
    graphPlotter.xticks(temporar, givenRanges, rotation='vertical')
    graphPlotter.legend(framealpha=1, frameon=True, loc='upper left')
    graphPlotter.title("Value of seed(x_0) = "+str(Seed))
    graphPlotter.ylabel("Frequency")
    graphPlotter.xlabel("Ranges")
    graphPlotter.show()
```

```python
# Question 3
import matplotlib.pyplot as graphPlotter
print("The Linear Congruence Generator is: ")
print("x_i+1 = (a * x_i + b) mod m")
print("u_i+1 = x_i+1 / m")
print()
print("Question 3 :")
print()
def helper():
    x0 = 1
    ui = list()
    xi = list()
    while True:
        x0 = (1229 * x0 + 1) % 2048
        u0 = x0/2048
        if x0 in xi:
            break
        elif(u0 in ui):
            break    
        if(x0 == 0):
            continue
        xi.append(x0)
        ui.append(u0)
    return ui

Sequence1 = helper()
Sequence2 = helper()
print("Length = ", len(Sequence1))
del Sequence1[len(Sequence1) - 1]
del Sequence2[0]
print("Sequence 1 = ", end=' ')
print(Sequence1)
print("Sequence 2 = ", end=' ')
print(Sequence2)
fig, ax = graphPlotter.subplots()
ax.scatter(Sequence1, Sequence2)
graphPlotter.title("Two Dimensional Graph of the Points (u_i-1 , u_i)")
graphPlotter.ylabel("Y-axis")
graphPlotter.xlabel("X-axis")
graphPlotter.show()
```

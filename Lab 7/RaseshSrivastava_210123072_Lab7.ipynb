```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 07

# Question 2
import numpy as npy
import pandas as pd

ValuesOfM_toBeTaken = [100, 1000, 10000, 100000]

def MonteCarloEstimator_Lab6(M):
    npy.random.seed(36)
    U = npy.random.uniform(0, 1, M)
    Y = npy.exp(npy.sqrt(U))
    I_M = npy.mean(Y)
    se = npy.std(Y) / npy.sqrt(M)
    The_95_percent_ConfidenceInterval = (I_M - 1.96 * se, I_M + 1.96 * se)
    LowerBound = I_M - 1.96 * se
    UpperBound = I_M + 1.96 * se
    return I_M, LowerBound, UpperBound

def Antithetic_Method(m):
    i = 0
    ValuesOfY = []
    I_M = 0
    m = m/2
    m = int(m)
    # Generating m values of Y
    for i in range(0, m):
        U = npy.random.rand()
        ValueOf_x = npy.exp(npy.sqrt(U))
        ValueOf_xBar = npy.exp(npy.sqrt(1-U))
        ValuesOfY.append((ValueOf_x+ValueOf_xBar)/2)
        I_M = I_M + ((ValueOf_x+ValueOf_xBar)/2)
    I_M = I_M/m

    s_anti = 0
    for i in range(m):
        s_anti = s_anti + ((ValuesOfY[i]-I_M)**2)
    # Variance of Y
    s_anti = s_anti/(m-1)
    s_anti = npy.sqrt(s_anti)
    # Z_Value for 95% Confidence Interval
    Z_Value = 1.96
    # 95% Confidence Interval
    LowerBound = I_M - Z_Value * (s_anti / npy.sqrt(m))
    UpperBound = I_M + Z_Value * (s_anti / npy.sqrt(m))
    return I_M, LowerBound, UpperBound

DataFrameOfTheObtainedResults = pd.DataFrame(columns=['M', 'I_M', 'I_M using Antithetic Method', '95% Confidence Interval for I','95% Confidence Interval for I using Antithetic Method', 'Ratio of the Widths'])
# Actual value of I = Integral from 0 to 1 of exp(sqrt(x)) dx
ActualValueOf_I = 2

for m in ValuesOfM_toBeTaken:
    I_M, LowerBound, UpperBound = MonteCarloEstimator_Lab6(m)
    I_m_anti, LowerBoundUsingAntithetics, UpperBoundUsingAntithetics = Antithetic_Method(m)
    ratio_of_widths = (UpperBound - LowerBound) / (UpperBoundUsingAntithetics - LowerBoundUsingAntithetics)
    new_row = {'M': m, 'I_M': I_M, 'I_M using Antithetic Method': I_m_anti, '95% Confidence Interval for I': (
        LowerBound, UpperBound), '95% Confidence Interval for I using Antithetic Method': (LowerBoundUsingAntithetics, UpperBoundUsingAntithetics), 'Ratio of the Widths': ratio_of_widths}
    DataFrameOfTheObtainedResults = pd.concat([DataFrameOfTheObtainedResults, pd.DataFrame([new_row])], ignore_index=True)

print("Comparison of the results of Question 1 of Lab 6 and Question 1 of Lab 7 in a tabular form:")
print(DataFrameOfTheObtainedResults)
print(f"\nActual value of I = {ActualValueOf_I}")
```

```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 11

# Question 1
import statistics as st
import math
import random

GeneratedValues=[]
Ï€ = [1/2, 1/3, 1/6]
meu=[-1,0,1]
Sigma=[1/4,1,1/2]
# loop is till n/2 because in each iteration, we get 2 random variables
for i in range(0,2500):
    # generating random variables with normal distribution using Box-Muller Method
    FirstUniformNumberGenerated=random.random()
    SecondUniformNumberGenerated=random.random()
    R=math.sqrt(-2*math.log(FirstUniformNumberGenerated))
    TheTheta=2*math.pi*SecondUniformNumberGenerated
    z1=R*math.cos(TheTheta)
    z2=R*math.sin(TheTheta)
    # generating a random number from uniform distribution which will determine which f to take
    ThirdUniformNumberGenerated=random.random()
    if ThirdUniformNumberGenerated<=1/2:
        GeneratedValues.append(meu[0]+Sigma[0]*z1)
    elif ThirdUniformNumberGenerated<=5/6:
        GeneratedValues.append(meu[1]+Sigma[1]*z1)
    else:
        GeneratedValues.append(meu[2]+Sigma[2]*z1)

    FourthUniformNumberGenerated=random.random()
    if FourthUniformNumberGenerated<=1/2:
        GeneratedValues.append(meu[0]+Sigma[0]*z2)
    elif FourthUniformNumberGenerated<=5/6:
        GeneratedValues.append(meu[1]+Sigma[1]*z2)
    else:
        GeneratedValues.append(meu[2]+Sigma[2]*z2)

print(f"The number of values generated is: {len(GeneratedValues)}")
print(f"The average of the generated random numbers is: {st.mean(GeneratedValues)}")
print(f"The variance of the generated random numbers is: {st.variance(GeneratedValues)}")
```

```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 11

# Question 2
import math
import random
import matplotlib.pyplot as GraphPlotter
GeneratedValues=[] #values[i] will keep track of the values of (i+1)th graph
for j in range(0,10):
    GeneratedValues.append([0])

time=[0] # this list will store the uniform time intervals on which value of W(t(i))s are computed
for i in range(0,5000):
    time.append(time[-1]+1/1000)

for j in range(0,10):
    for i in range(0,2500):
        # generating random variables with normal distribution using Box-Muller Method
        FirstUniformNumberGenerated=random.random()
        SecondUniformNumberGenerated=random.random()
        R=math.sqrt(-2*math.log(FirstUniformNumberGenerated))
        TheTheta=2*math.pi*SecondUniformNumberGenerated
        z1=R*math.cos(TheTheta)
        z2=R*math.sin(TheTheta)
        # value generated from the formula given in the question, once for z1 and then for z2
        w=GeneratedValues[j][-1]+math.sqrt(1/1000)*z1
        GeneratedValues[j].append(w)
        w=GeneratedValues[j][-1]+math.sqrt(1/1000)*z2
        GeneratedValues[j].append(w)

for j in range(0,10):
    GraphPlotter.plot(time, GeneratedValues[j], label=str(j+1), linewidth=0.7)

GraphPlotter.xlabel('Time')
GraphPlotter.ylabel('W(t_i)')
GraphPlotter.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
GraphPlotter.title('Sample Paths of Standard Brownian Motion: W(t) vs t')
GraphPlotter.show()

AverageValueOf2=0
AverageValueOf5=0
# calculating the expectation of W(2) and W(5)
for abc in range(0,10):
    AverageValueOf2 = AverageValueOf2+GeneratedValues[abc][2000]
    AverageValueOf5 = AverageValueOf5+GeneratedValues[abc][5000]

AverageValueOf2=AverageValueOf2/10
AverageValueOf5=AverageValueOf5/10
print(f"The number of values generated for each graph is: {len(GeneratedValues[0])-1}")
# print(f"The Actual Value of E(W(2)) is: {0}")
print(f"The Estimated Value of E(W(2)) is: {AverageValueOf2}")
# print(f"The Actual Value of E(W(5)) is: {0}")
print(f"The Estimated Value of E(W(5)) is: {AverageValueOf5}")
```

```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 11

# Question 3
import math
import random
import matplotlib.pyplot as GraphPlotter
meu=0.06
Sigma=0.3
GeneratedValues=[] # list to store the generated values, values[i] will keep track of the values of (i+1)th graph
for j in range(0,10):
    GeneratedValues.append([5])

time=[0] # this list will store the uniform time intervals on which value of X(t(i))s are computed
for i in range(0,5000):
    time.append(time[-1]+1/1000)

for j in range(0,10):
    for i in range(0,2500):
        # generating random variables with normal distribution using Box-Muller Method
        FirstUniformNumberGenerated=random.random()
        SecondUniformNumberGenerated=random.random()
        R=math.sqrt(-2*math.log(FirstUniformNumberGenerated))
        TheTheta=2*math.pi*SecondUniformNumberGenerated
        z1=R*math.cos(TheTheta)
        z2=R*math.sin(TheTheta)
        # value generated from the formula given in the question, once for z1 and then for z2
        x=GeneratedValues[j][-1]+meu*(1/1000)+Sigma*math.sqrt(1/1000)*z1
        GeneratedValues[j].append(x)
        x=GeneratedValues[j][-1]+meu*(1/1000)+Sigma*math.sqrt(1/1000)*z2
        GeneratedValues[j].append(x)

for abc in range(0,10):
    GraphPlotter.plot(time, GeneratedValues[abc], label=str(abc+1), linewidth=0.7)

GraphPlotter.xlabel('Time')
GraphPlotter.ylabel('X(t_i)')
GraphPlotter.title('Sample Paths of Brownian Motion with Drift: X(t) vs t')
GraphPlotter.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
GraphPlotter.show()
AverageValueOf2=0
AverageValueOf5=0
# calculating the expectation of X(2) and X(5)
for abc in range(0,10):
    AverageValueOf2 = AverageValueOf2+GeneratedValues[abc][2000]
    AverageValueOf5 = AverageValueOf5+GeneratedValues[abc][5000]

AverageValueOf2=AverageValueOf2/10
AverageValueOf5=AverageValueOf5/10
print(f"The number of values generated for each graph is: {len(GeneratedValues[0])-1}")
print(f"The Estimated Value of E(X(2)) is: {AverageValueOf2}")
print(f"The Estimated Value of E(X(5)) is: {AverageValueOf5}")
```

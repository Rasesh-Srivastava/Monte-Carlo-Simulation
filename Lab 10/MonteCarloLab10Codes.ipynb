```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 10

# Question 2
import matplotlib.pyplot as GraphPlotter

# list to store the powers of 2
PowerOf2=[]
The2=1
for i in range(0,19):
    PowerOf2.append(The2)
    The2*=2

# function to generate Van Der Corput Sequence in base 2
def VanDerCorputSequenceGenerator(n):
    x=0
    y=n
    for i in range(18,-1,-1):
        if y>=PowerOf2[i]:
            y-=PowerOf2[i]
            x+=1/(PowerOf2[i]*2)
    return x

def BothTheMethodsGenerator(n):
    VanDerCoruptValues=[]
    for i in range(1,n+1):
        VanDerCoruptValues.append(VanDerCorputSequenceGenerator(i))

    LinearCongruenceGeneratedValues=[]
    m=1000000007
    a=12
    c=8
    Seed=3
    for i in range(0,n):
        Seed=(a*Seed+c)%m
        LinearCongruenceGeneratedValues.append(Seed/m)

    # Plotting the histograms
    GraphPlotter.title(f"Histogram for n = {n} with bins = 20")
    GraphPlotter.hist(VanDerCoruptValues, label='Van Der Corput',bins=20,color='red')
    GraphPlotter.xlabel("Values")
    GraphPlotter.ylabel("Frequency")
    GraphPlotter.legend(loc='lower right')
    GraphPlotter.show()

    GraphPlotter.title(f"Histogram for n = {n} with bins = 20")
    GraphPlotter.hist(LinearCongruenceGeneratedValues, label='LCG', bins=20,color='green')
    GraphPlotter.xlabel("Values")
    GraphPlotter.ylabel("Frequency")
    GraphPlotter.legend(loc='lower right')
    GraphPlotter.show()

    GraphPlotter.title("Comparing the Histograms for n = "+str(n))
    GraphPlotter.hist([VanDerCoruptValues,LinearCongruenceGeneratedValues], label=['Van Der Corput', 'LCG'],color=['red','green'])
    GraphPlotter.xlabel("Values")
    GraphPlotter.ylabel("Frequency")
    GraphPlotter.legend(loc='lower right')
    GraphPlotter.show()

# calling the function for n = 100
BothTheMethodsGenerator(100)
# calling the function for n = 100000
BothTheMethodsGenerator(100000)
```

```python
# Rasesh Srivastava
# Roll Number - 210123072

# MA323 Monte Carlo Simulation
# Lab 10

# Question 3
import matplotlib.pyplot as GraphPlotter

PowersOf3=[]
The3=1
for i in range(0,12):
    PowersOf3.append(The3)
    The3*=3

# function to generate Van Der Corput Sequence in base 3
def VanDerCorputSequenceGenerator_3(n):
    x=0
    y=n
    for i in range(11,-1,-1):
        if y>=PowersOf3[i]:
            z=y//PowersOf3[i]
            y%=PowersOf3[i]
            x+=z/(PowersOf3[i]*3)
    return x

PowerOf2=[]
The2=1
for i in range(0,19):
    PowerOf2.append(The2)
    The2*=2

# function to generate Van Der Corput Sequence in base 2
def VanDerCorputSequenceGenerator_2(n):
    x=0
    y=n
    for i in range(18,-1,-1):
        if y>=PowerOf2[i]:
            y-=PowerOf2[i]
            x+=1/(PowerOf2[i]*2)
    return x

def HaltonSequenceGenerator(n):
    v2=[]
    v3=[]
    for i in range(1,n+1):
        v2.append(VanDerCorputSequenceGenerator_2(i))
        v3.append(VanDerCorputSequenceGenerator_3(i))

    GraphPlotter.scatter(v2, v3)
    GraphPlotter.xlabel("Phi_2(i)")
    GraphPlotter.ylabel("Phi_3(i)")
    GraphPlotter.title(f"Scatter plot of Halton Sequence for n = {n}")
    GraphPlotter.show()

# calling the function for n = 100
HaltonSequenceGenerator(100)
# calling the function for n = 100000
HaltonSequenceGenerator(100000)
```

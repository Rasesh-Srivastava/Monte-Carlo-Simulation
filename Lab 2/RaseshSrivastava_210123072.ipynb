```python
# MA 323  Monte Carlo Simulation 
# Lab 2

# Question 1
import matplotlib.pyplot as GraphPlotter
import numpy as np

def LinearCongruenceGenerator(a, b, m, seed):
  xi = list()
  ui = list()
  x = seed
  while True:
      x = (a * x + b) % m
      u = x/m
      if x in xi or u in ui:
          break

      if(x == 0):
          continue
      xi.append(x)
      ui.append(u)

  return ui

def RecursiveGenerator(limit, U):
  limit = limit -  17
  i = len(U)

  for index in range(1, limit + 1):
    rand = U[i - 17] - U[i - 5]
    
    if rand < 0:
      rand = rand + 1

    U.append(rand)
    i = i + 1

  return U

# For N = 1000
list1 = LinearCongruenceGenerator(1229, 1, 2048, 1)
list2 = LinearCongruenceGenerator(1229, 1, 2048, 1)
x = RecursiveGenerator(1000, list1[0: 17])
y = RecursiveGenerator(1000, list2[0: 17])

GraphPlotter.hist(x, bins=np.arange(min(x), max(x)+0.001, 0.01),color='#CA3433')
GraphPlotter.ylabel("Frequency")
GraphPlotter.xlabel("U_i")
GraphPlotter.title("Histogram: Frequency Distribution plot for N = 1000")
GraphPlotter.show()

del x[len(x) - 1]
del y[0]

fig, ax = GraphPlotter.subplots()
ax.scatter(x, y, alpha=0.5)
GraphPlotter.ylabel("Y axis")
GraphPlotter.xlabel("X axis")
GraphPlotter.title("Scatter plot (U_i, U_i+1) for N = 1000")
GraphPlotter.show()



# For N = 10000
list1 = LinearCongruenceGenerator(1229, 1, 2048, 1)
list2 = LinearCongruenceGenerator(1229, 1, 2048, 1)
x = RecursiveGenerator(10000, list1[0: 17])
y = RecursiveGenerator(10000, list2[0: 17])

GraphPlotter.hist(x, bins=np.arange(min(x), max(x)+0.001, 0.01),color='#CA3433')
GraphPlotter.ylabel("Frequency")
GraphPlotter.xlabel("U_i")
GraphPlotter.title("Histogram: Frequency Distribution plot for N = 10000")

GraphPlotter.show()


del x[len(x) - 1]
del y[0]

fig, ax = GraphPlotter.subplots()
ax.scatter(x, y, alpha=0.5)
GraphPlotter.title("Scatter plot (U_i, U_i+1) for N = 10000")
GraphPlotter.ylabel("Y axis")
GraphPlotter.xlabel("X axis")
GraphPlotter.show()



# For N = 100000
list1 = LinearCongruenceGenerator(1229, 1, 2048, 1)
list2 = LinearCongruenceGenerator(1229, 1, 2048, 1)
x = RecursiveGenerator(100000, list1[0: 17])
y = RecursiveGenerator(100000, list2[0: 17])

GraphPlotter.hist(x, bins=np.arange(min(x), max(x)+0.001, 0.01),color='#CA3433')
GraphPlotter.ylabel("Frequency")
GraphPlotter.xlabel("U_i")
GraphPlotter.title("Histogram: Frequency Distribution plot for N = 100000")
GraphPlotter.show()

del x[len(x) - 1]
del y[0]

fig, ax = GraphPlotter.subplots()
ax.scatter(x, y, alpha=0.5)
GraphPlotter.title("Scatter plot (U_i, U_i+1) for N = 100000")
GraphPlotter.ylabel("Y axis")
GraphPlotter.xlabel("X axis")
GraphPlotter.show()

```

```python
# Question 2
import math 
import matplotlib.pyplot as GraphPlotter
import numpy as np
from numpy import random
import cmath 

Different_N_Values = [10, 100, 1000, 10000, 100000]
for N in Different_N_Values:
  X = list()
  ta = cmath.pi/3
  np.random.seed(42)
  GeneratedSeq = random.uniform(size=N) 
  l = len(GeneratedSeq)
  print('For N = {},'.format(l))

  for j in GeneratedSeq:
    val = -ta * math.log(1 - j)
    X.append(val)

  Y_final = np.sort(X)
  lY = len(Y_final)
  valueOf_Y = np.arange(lY)/float(lY - 1)
  
  GraphPlotter.ylabel('Probability of (X <= x)')
  GraphPlotter.xlabel('X (the generated values)')
  GraphPlotter.title('Empirical CDF: Distribution Function of X (For N = {})'.format(N))
  GraphPlotter.plot(Y_final, valueOf_Y, color='#CA3433')
  GraphPlotter.show()

  print("Sample count (N) =", N,end="")
  print(", Sample Mean =", np.mean(X),end ="")
  print(", Sample Variance =", np.var(X))
  print()

temp = np.arange(0, 25, 0.01)
FunctionOfX = [1 - math.exp(-j/ta) for j in temp]

GraphPlotter.plot(temp, FunctionOfX, color='green')
GraphPlotter.ylabel('F(x)')
GraphPlotter.xlabel('x')
GraphPlotter.title('Actual Distribution Function (CDF)')
GraphPlotter.show()

print("Actual Mean =", ta)
print("Actual Variance =", ta*ta)
```

```python
# Question 3
import math 
import matplotlib.pyplot as GraphPlotter
import numpy as np
from numpy import random
import cmath 

Different_N_Values = [10, 100, 1000, 10000, 100000]
for N in Different_N_Values:
  X = list()
  np.random.seed(42)
  GeneratedSeq = random.uniform(size=N) 
  l = len(GeneratedSeq)
  print('For N = {},'.format(l))

  for j in GeneratedSeq:
    val = 0.5 - 0.5*math.cos(j*cmath.pi)
    X.append(val)

  Y_final = np.sort(X)
  lY = len(Y_final)
  valueOf_Y = np.arange(lY)/float(lY - 1)
  
  GraphPlotter.ylabel('Probability of (X <= x)')
  GraphPlotter.xlabel('X (the generated values)')
  GraphPlotter.title('Empirical CDF: Distribution Function of X (For N = {})'.format(N))
  GraphPlotter.plot(Y_final, valueOf_Y, color='#CA3433')
  GraphPlotter.show()

  print("Sample count (N) =", N,end="")
  print(", Sample Mean =", np.mean(X),end ="")
  print(", Sample Variance =", np.var(X))
  print()

temp = np.arange(0, 1.001, 0.01)
FunctionOfX = [(2/cmath.pi)*math.asin(math.sqrt(j)) for j in temp]

GraphPlotter.plot(temp, FunctionOfX, color='green')
GraphPlotter.ylabel('F(x)')
GraphPlotter.xlabel('x')
GraphPlotter.title('Actual Distribution Function (CDF)')
GraphPlotter.show()
```

```python
# Question 4
from numpy import random
import matplotlib.pyplot as GraphPlotter
import numpy as np
import math 

Sequence = [i for i in range(1, 10000, 2)]
prob = [i/5000 for i in range(1, 5001)]
np.random.seed(29)
U = random.uniform(size = 100000) 
SampleList = list()    # List of the generated random numbers X
for ui in U:
    beginning = 0
    last = len(prob) - 1
    while beginning < last:
        middle = math.floor((beginning + last)/2)
        if(ui <= prob[middle]):
            last = middle
        else:
            beginning = middle + 1
    SampleList.append(Sequence[beginning - 1])
    
hist, bins = np.histogram(SampleList,bins=5000)
w = 0.73 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
GraphPlotter.bar(center, hist, align = 'center', width = w, color='#CA3433')
GraphPlotter.title('Frequency distribution of X (For N = 100000)')
GraphPlotter.ylabel('Frequency')
GraphPlotter.xlabel('Random Numbers (X)')
GraphPlotter.show()
```
